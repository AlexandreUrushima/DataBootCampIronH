```python
#Rolling the dice game
import random

#Globals!
Money = 1000
min = 1
max = 6
act_bet = 0

def oneinsix(current_bet):
    '''
     Enter the value of a bet as arg. A function to try your lucky guessing a random number in a dice.
    '''
    global Money
    global act_bet
    user_input = int(input('Input a number between 1 and 6: ', ))
    if (user_input < 1) or (user_input > 6):
        user_input = input('Please give a number between 1 and 6: ', )
    print ("Rolling the dices...")
    print ("The values are....")
    cpu_input = random.randint(min, max)
    print ('Your number is: ',user_input,'Dices number is: ',cpu_input)
    if user_input == cpu_input:
        act_bet = current_bet * 2
        print('Money earned: ', act_bet)
        print ("User wins!")
    else:
        act_bet = 0
        print('Money earned: ', act_bet)
        print ("Cpu wins!")
    return act_bet

def evenorodd(current_bet):
    '''
     Enter the value of a bet as arg. A function to try your lucky guessing even or odd with a dice.
    '''
    global Money
    global act_bet
    user_input = int(input('Input 1 for pairs or 2 for odds', ))
    if (user_input != 1) and (user_input != 2):
        user_input = input('Please input 1 or 2: ', )
    print ("Rolling the dices...")
    print ("The values are....")
    cpu_input = random.randint(min, max)
    print ('Your number is: ',user_input,'Dices number is: ',cpu_input)
    if (user_input % 2) == 0:
        act_bet = current_bet * 1.5
        print('Money earned: ', current_bet)
        print ("User wins!")
    else:
        act_bet = 0
        print('Money earned: ', current_bet)
        print ("Cpu wins!")
        
    return act_bet

def main(): #MAIN GAME LOOP
    print('\nWelcome to Dice Game, and prepare do earn easy money.\n')
    global Money ## retirar variavel global
    print(f' This is your current Money : {Money}')
    while Money > 0:
        game_select = int(input('Input 1 for One in Six Game and win 2x your stake or Input 2 for Even or Odd and win 1.5x: ', )) ## corrigir para números negativos
        current_bet = int(input('Input your bet to the game: ', )) ## corrigir para números negativos e não quebrar quando for diferente de um int
        if current_bet > Money:
            while current_bet > Money:
                current_bet = int(input('You dont have this money! Input a value that you have: ', )) ## corrigir para números negativos
        if (game_select == 1):
            Money -= current_bet
            oneinsix(current_bet)
            Money += act_bet
            print(f' This is your current Money : {Money}')
        elif (game_select == 2):
            Money -= current_bet
            evenorodd(current_bet)
            Money += act_bet
            print(f' This is your current Money : {Money}')
        while True:
            answer = input('Do you want to reset? (Y/N)', )
            if answer.lower() == 'y' or answer.lower == 'yes':
                main()
            else:
                break
        print('-'*100)

main()

```
